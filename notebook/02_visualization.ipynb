```python
# Import packages
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import genfromtxt
```

```python
# Take one figure as an example
data = genfromtxt('../data/processed/2022-08-03/view_1/snap-2022-08-03T13-48-18.19.csv', delimiter=',')
print(f'Data shape: {data.shape}')
```

```python
# Visualize the figure
fig, ax = plt.subplots(1, figsize = (6.4, 4.8))
im = ax.imshow(data, cmap=cm.Spectral_r)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
fig.subplots_adjust(right=0.9)
position = fig.add_axes([0.95, 0.11, 0.03, 0.77]) # [left, down, right, up]
cb = fig.colorbar(im, cax=position)
cb.ax.set_ylabel('Temperature in Degrees Celcius', fontsize=13)
cb.ax.tick_params(labelsize=10)
```
